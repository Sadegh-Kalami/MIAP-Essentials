```python
# Imports
import os
import sys
import random
import itertools
import numpy as np
import cv2
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
import pandas as pd
import datetime
import seaborn as sns

# Seeding for reproducibility
np.random.seed = 42
tf.random.set_seed(42)

# Data Generator
class DataGen(keras.utils.Sequence):
    def __init__(self, ids, path, batch_size=8, image_size=128, shuffle=True):
        self.ids = ids
        self.path = path
        self.batch_size = batch_size
        self.image_size = image_size
        self.shuffle = shuffle
        self.on_epoch_end()

    def __load__(self, id_name):
        # Paths
        image_path = os.path.join(self.path, id_name, "images", id_name) + ".png"
        mask_path = os.path.join(self.path, id_name, "masks/")
        all_masks = os.listdir(mask_path)

        # Reading Image
        image = cv2.imread(image_path, 1)
        image = cv2.resize(image, (self.image_size, self.image_size))

        mask = np.zeros((self.image_size, self.image_size, 1))

        # Reading Masks
        for name in all_masks:
            _mask_path = os.path.join(mask_path, name)
            _mask_image = cv2.imread(_mask_path, -1)
            _mask_image = cv2.resize(_mask_image, (self.image_size, self.image_size))
            _mask_image = np.expand_dims(_mask_image, axis=-1)
            mask = np.maximum(mask, _mask_image)

        # Normalizing
        image = image / 255.0
        mask = mask / 255.0

        return image, mask

    def __getitem__(self, index):
        batch_ids = self.ids[index * self.batch_size:(index + 1) * self.batch_size]
        image = []
        mask = []

        for id_name in batch_ids:
            _img, _mask = self.__load__(id_name)
            image.append(_img)
            mask.append(_mask)

        image = np.array(image)
        mask = np.array(mask)

        return image, mask

    def on_epoch_end(self):
        if self.shuffle:
            random.shuffle(self.ids)

    def __len__(self):
        return int(np.ceil(len(self.ids) / float(self.batch_size)))

# U-Net Model Definition
def down_block(x, filters, kernel_size=(3, 3), padding="same", strides=1):
    c = keras.layers.Conv2D(filters, kernel_size, padding=padding, strides=strides, activation="relu")(x)
    c = keras.layers.Conv2D(filters, kernel_size, padding=padding, strides=strides, activation="relu")(c)
    p = keras.layers.MaxPool2D((2, 2), (2, 2))(c)
    return c, p

def up_block(x, skip, filters, kernel_size=(3, 3), padding="same", strides=1):
    us = keras.layers.UpSampling2D((2, 2))(x)
    concat = keras.layers.Concatenate()([us, skip])
    c = keras.layers.Conv2D(filters, kernel_size, padding=padding, strides=strides, activation="relu")(concat)
    c = keras.layers.Conv2D(filters, kernel_size, padding=padding, strides=strides, activation="relu")(c)
    return c

def bottleneck(x, filters, kernel_size=(3, 3), padding="same", strides=1):
    c = keras.layers.Conv2D(filters, kernel_size, padding=padding, strides=strides, activation="relu")(x)
    c = keras.layers.Conv2D(filters, kernel_size, padding=padding, strides=strides, activation="relu")(c)
    return c

def UNet(filters=[16, 32, 64, 128, 256], input_size=(128, 128, 3)):
    inputs = keras.layers.Input(input_size)

    p0 = inputs
    c1, p1 = down_block(p0, filters[0])  # 128 --> 64
    c2, p2 = down_block(p1, filters[1])  # 64  --> 32
    c3, p3 = down_block(p2, filters[2])  # 32  --> 16
    c4, p4 = down_block(p3, filters[3])  # 16  --> 8

    bn = bottleneck(p4, filters[4])

    u1 = up_block(bn, c4, filters[3])  # 8  --> 16
    u2 = up_block(u1, c3, filters[2])  # 16 --> 32
    u3 = up_block(u2, c2, filters[1])  # 32 --> 64
    u4 = up_block(u3, c1, filters[0])  # 64 --> 128

    outputs = keras.layers.Conv2D(1, (1, 1), padding="same", activation="sigmoid")(u4)
    model = keras.models.Model(inputs, outputs)
    return model

# Hyperparameter options
epochs_options = [5, 10, 20]
batch_size_options = [2, 8, 16]
learning_rate_options = [0.0001, 0.001, 0.01]
filters_options = [
    [8, 16, 32, 64, 128],    # Reduced Capacity
    [16, 32, 64, 128, 256],  # Baseline Configuration
    [32, 64, 128, 256, 512]  # Increased Capacity
]

# Paths and Settings
image_size = 128
train_path = "stage1_train/"  # Ensure this path is correct
val_data_size = 55

# Get Training and Validation IDs
all_ids = next(os.walk(train_path))[1]
valid_ids = all_ids[:val_data_size]
train_ids = all_ids[val_data_size:]

# Create a timestamp for logging
timestamp = datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
log_file = f"hyperparameter_tuning_log_{timestamp}.csv"

# Initialize a list to store results
results = []

# Iterate through all combinations
total_combinations = len(epochs_options) * len(batch_size_options) * len(learning_rate_options) * len(filters_options)
current_combination = 1

for epochs, batch_size, lr, filters in itertools.product(
    epochs_options, batch_size_options, learning_rate_options, filters_options
):
    print(f"\nTraining combination {current_combination}/{total_combinations}")
    print(f"Epochs: {epochs}, Batch Size: {batch_size}, Learning Rate: {lr}, Filters: {filters}")

    # Define Model
    model = UNet(filters=filters, input_size=(image_size, image_size, 3))
    optimizer = keras.optimizers.Adam(learning_rate=lr)
    model.compile(optimizer=optimizer, loss="binary_crossentropy", metrics=["accuracy"])

    # Define Callbacks
    model_dir = "models"
    os.makedirs(model_dir, exist_ok=True)
    # Create a unique filename for each combination
    filters_str = "_".join(map(str, filters))
    model_filename = f"UNet_e{epochs}_bs{batch_size}_lr{lr}_f{filters_str}.keras"
    checkpoint_path = os.path.join(model_dir, model_filename)
    
    callbacks = [
        EarlyStopping(monitor='val_accuracy', patience=5, verbose=1, restore_best_weights=True),
        ModelCheckpoint(checkpoint_path, monitor='val_accuracy', save_best_only=True, verbose=1)
    ]

    # Create Data Generators
    train_gen = DataGen(train_ids, train_path, image_size=image_size, batch_size=batch_size, shuffle=True)
    valid_gen = DataGen(valid_ids, train_path, image_size=image_size, batch_size=batch_size, shuffle=False)

    # Calculate Steps
    train_steps = len(train_ids) // batch_size
    valid_steps = len(valid_ids) // batch_size

    # Train the Model
    history = model.fit(
        train_gen,
        validation_data=valid_gen,
        steps_per_epoch=train_steps,
        validation_steps=valid_steps,
        epochs=epochs,
        callbacks=callbacks,
        verbose=1
    )

    # Extract the best validation accuracy
    best_val_accuracy = max(history.history['val_accuracy'])

    # Log the results
    result = {
        'Epochs': epochs,
        'Batch_Size': batch_size,
        'Learning_Rate': lr,
        'Filters': filters,
        'Best_Val_Accuracy': best_val_accuracy,
        'Model_Path': checkpoint_path
    }
    results.append(result)

    print(f"Completed combination {current_combination}/{total_combinations} - Best Val Accuracy: {best_val_accuracy}")
    current_combination += 1

# Save the results to a CSV file
df_results = pd.DataFrame(results)
df_results.to_csv(log_file, index=False)
print(f"\nHyperparameter tuning completed. Results saved to {log_file}")

# Load the results
df_results = pd.read_csv(log_file)

# Display the results
print("\nHyperparameter Tuning Results:")
print(df_results)

# Plotting Validation Accuracy vs. Hyperparameters

# Set the aesthetic style of the plots
sns.set(style="whitegrid")

# Plot Validation Accuracy for different Epochs
plt.figure(figsize=(12, 6))
sns.boxplot(x='Epochs', y='Best_Val_Accuracy', data=df_results)
plt.title('Validation Accuracy vs. Number of Epochs')
plt.show()

# Plot Validation Accuracy for different Batch Sizes
plt.figure(figsize=(12, 6))
sns.boxplot(x='Batch_Size', y='Best_Val_Accuracy', data=df_results)
plt.title('Validation Accuracy vs. Batch Size')
plt.show()

# Plot Validation Accuracy for different Learning Rates
plt.figure(figsize=(12, 6))
sns.boxplot(x='Learning_Rate', y='Best_Val_Accuracy', data=df_results)
plt.title('Validation Accuracy vs. Learning Rate')
plt.show()

# Plot Validation Accuracy for different Number of Filters
plt.figure(figsize=(12, 6))
# Convert filters list to string for better visualization
df_results['Filters_Str'] = df_results['Filters'].apply(lambda x: str(x))
sns.boxplot(x='Filters_Str', y='Best_Val_Accuracy', data=df_results)
plt.xticks(rotation=45)
plt.title('Validation Accuracy vs. Number of Filters')
plt.show()

# Identify the best model
best_row = df_results.loc[df_results['Best_Val_Accuracy'].idxmax()]
best_model_path = best_row['Model_Path']
best_accuracy = best_row['Best_Val_Accuracy']
best_hyperparams = {
    'Epochs': best_row['Epochs'],
    'Batch_Size': best_row['Batch_Size'],
    'Learning_Rate': best_row['Learning_Rate'],
    'Filters': eval(best_row['Filters']) if isinstance(best_row['Filters'], str) else best_row['Filters']
}

print("\nBest Model Configuration:")
print(f"Epochs: {best_hyperparams['Epochs']}")
print(f"Batch Size: {best_hyperparams['Batch_Size']}")
print(f"Learning Rate: {best_hyperparams['Learning_Rate']}")
print(f"Filters: {best_hyperparams['Filters']}")
print(f"Validation Accuracy: {best_accuracy}")
print(f"Model Saved at: {best_model_path}")

# Load the best model
best_model = UNet(filters=best_hyperparams['Filters'], input_size=(image_size, image_size, 3))
best_model.compile(optimizer=keras.optimizers.Adam(learning_rate=best_hyperparams['Learning_Rate']),
                   loss="binary_crossentropy", metrics=["accuracy"])
best_model.load_weights(best_model_path)

# Save the best model in a separate directory if desired
best_model_dir = "best_model"
os.makedirs(best_model_dir, exist_ok=True)
best_model.save(os.path.join(best_model_dir, "UNet_best_model.keras"))
print(f"Best model saved at {os.path.join(best_model_dir, 'UNet_best_model.keras')}")

# Prediction and Visualization with the Best Model

# Select a batch from the validation set
selected_batch = 0  # Change as needed
valid_gen = DataGen(valid_ids, train_path, image_size=image_size, batch_size=best_hyperparams['Batch_Size'], shuffle=False)
x_val, y_val = valid_gen.__getitem__(selected_batch)

# Make predictions
predictions = best_model.predict(x_val)
predictions = predictions > 0.5  # Binarize the output

# Plot some predictions
num_samples = min(4, len(x_val))  # Plot up to 4 samples
plt.figure(figsize=(12, num_samples * 4))

for i in range(num_samples):
    # Original Image
    plt.subplot(num_samples, 3, i * 3 + 1)
    plt.imshow(x_val[i])
    plt.title("Original Image")
    plt.axis('off')

    # Ground Truth Mask
    plt.subplot(num_samples, 3, i * 3 + 2)
    plt.imshow(y_val[i].reshape(image_size, image_size), cmap="gray")
    plt.title("Ground Truth Mask")
    plt.axis('off')

    # Predicted Mask
    plt.subplot(num_samples, 3, i * 3 + 3)
    plt.imshow(predictions[i].reshape(image_size, image_size), cmap="gray")
    plt.title("Predicted Mask")
    plt.axis('off')

plt.tight_layout()
plt.show()
```
